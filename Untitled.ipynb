* 출처: https://github.com/ssooni/data_mining_practice

```python
import re
import pandas as pd
import datetime as dt

def read_kko_msg(filename):
    with open(filename, encoding='utf-8') as f:
        msg_list = f.readlines()
    return msg_list

def apply_kko_regex(msg_list):
    kko_pattern = re.compile("\[([\S\s]+)\] \[(오전|오후) ([0-9:\s]+)\] ([^\n]+)")
    kko_date_pattern = re.compile("--------------- ([0-9]+년 [0-9]+월 [0-9]+일) ")

    emoji_pattern = re.compile("["u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               "]+", flags=re.UNICODE)

    kko_parse_result = list()
    cur_date = ""

    for msg in msg_list:
        # 날짜 부분인 경우
        if len(kko_date_pattern.findall(msg)) > 0:
            cur_date = dt.datetime.strptime(kko_date_pattern.findall(msg)[0], "%Y년 %m월 %d일")
            cur_date = cur_date.strftime("%Y-%m-%d")
        else:
            kko_pattern_result = kko_pattern.findall(msg)
            if len(kko_pattern_result) > 0:
                tokens = list(kko_pattern_result[0])
                # 이모지 데이터 삭제
                tokens[-1] = re.sub(emoji_pattern, "", tokens[-1])
                tokens.insert(0, cur_date)
                kko_parse_result.append(tokens)

    kko_parse_result = pd.DataFrame(kko_parse_result, columns=["Date", "Speaker", "timetype", "time", "contents"])
    kko_parse_result.to_csv("C:/Users/user/github/kakao/data/kko_regex.csv", index=False)

    return kko_parse_result

if __name__ == '__main__':
    msg_list = read_kko_msg("C:/Users/user/github/kakao/data/카톡대화.txt")
    apply_kko_regex(msg_list)
    
df = pd.read_csv("C:/Users/user/github/kakao/data/kko_regex.csv")
df.head()
```

### 년월일 분리, 요일 추가

* [pandas to date_time 참고](https://m.blog.naver.com/PostView.nhn?blogId=wideeyed&logNo=221603462366&proxyReferer=https:%2F%2Fwww.google.com%2F)

```python
# Date열의 dtype을 object series로부터 datetime64로 변환
df.Date = pd.to_datetime(df.Date, format='%Y-%m-%d', errors='raise')
df.Date.head()
```

```python
df['Year'] = df.Date.dt.year
df['Month'] = df.Date.dt.month
df['Day'] = df.Date.dt.day
df['Weekday'] = df.Date.dt.strftime("%a")
df.head()
```

### 24시간 표시

```python
temp = df['timetype']
temp = temp.apply(lambda x : x.replace('오전','AM'))
temp = temp.apply(lambda x : x.replace('오후','PM'))
```

```python
df['time24'] = temp + " " + df.time
df.time24.head()
```

```python
time_temp = []
for i in range(len(df)):
    time = dt.datetime.strptime(df.time24[i], '%p %I:%M')
    time_temp.append(time.strftime('%H:%M'))
df.time24 = time_temp
```

```python
h_temp=[]
m_temp=[]
for i in range(len(df)):
    hour = df.time24[i].split(':')[0]
    minute = df.time24[i].split(':')[1]
    h_temp.append(hour)
    m_temp.append(minute)
df['hour'] = h_temp
df['minute'] = m_temp
```

```python
df.dtypes
```

```python
# 시간 열을 numeric(int)로 변형
df.Year = df.Year.apply(pd.to_numeric)
df.hour = df.hour.apply(pd.to_numeric)
df.minute = df.minute.apply(pd.to_numeric)
```

```python
df.head()
```

### 글자수 추가

```python
len_temp = []
for i in range(len(df)):
    length = len(str(df.contents[i]))
    len_temp.append(length)
df['length'] = len_temp
```
